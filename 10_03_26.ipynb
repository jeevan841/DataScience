```python
import numpy as np
data = [10,20,30,40,50]
mean = np.mean(data)
print("Mean:", mean)
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
z_score = [(x - mean) / std_dev for x in data]
print("z_score:",z_score)
```

```python
import numpy as np
import math

population = np.array([10,20,30,40,50,60,70])
N = len(population)
print("population size (N):", N)

mu = np.mean(population)
print("population mean (μ):", mu)

sigma2 = np.var(population, ddof=0)
print("population variance (σ^2):", sigma2)

sigma = np.sqrt(sigma2)
print("population standard deviation:", sigma)

sample = np.array([20,30,40,50])
n = len(sample)
print("sample size (n):", n)

x_bar = np.mean(sample)
print("sample mean (x̄):", x_bar)

s2 = np.var(sample, ddof=1)
print("sample variance (s^2):", s2)

s = np.sqrt(s2)
print("sample standard deviation (s):", s)

SE = sigma / math.sqrt(n)
print("standard error (σ/√n):", SE)

z = (x_bar - mu) / SE
print("z-score:", z)
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 8))

ax.scatter(np.arange(len(sample)), sample, color='blue', label='Sample Points', s=100, zorder=5)

ax.axhline(y=x_bar, color='red', linestyle='--', label=f'Sample Mean (x̄ = {x_bar:.2f})')

ax.axhspan(x_bar - s, x_bar + s, color='green', alpha=0.2, label=f'1 Sample Std Dev (s = {s:.2f})')

ax.set_xlabel('Sample Index')
ax.set_ylabel('Value')
ax.set_title('Sample Data vs. Mean and Standard Deviation')
ax.set_xticks(np.arange(len(sample)))
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
