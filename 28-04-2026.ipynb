```python
import numpy as np
from scipy.spatial import distance
from scipy.stats import pearsonr
point1= np.array([2,3])
point2= np.array([5,7])
euclidean = distance.euclidean(point1,point2)
manhattan= distance.cityblock(point1,point2)
chebyshev= distance.chebyshev(point1,point2)
hamming= distance.hamming(point1,point2)
cosine_distance= distance.cosine(point1,point2)
cosine_similarity = 1-cosine_distance
set1,set2 = set(point1),set(point2)
jaccard_similarity = len(set1&set2)/len(set1|set2)
r,_= pearsonr(point1,point2)
print("euclidean distance:",euclidean)
print("manhattan distance:",manhattan)
print("chebyshev distance:",chebyshev)
print("hamming distance:",hamming)
print("cosine similarity:",cosine_similarity)
print("jaccard similarity:",jaccard_similarity)
print("pearsonr correlation:",r)
```

```python
import numpy as np
A= np.array([[1,2],[2,3],[3,4],[4,5]])
B= np.array([[1,4],[2,5],[3,6],[10,8]])
P= np.array([6,6])
CA = A.mean(axis= 0)
CB = B.mean(axis= 0)
DA = np.linalg.norm(P-CA)
DB = np.linalg.norm(P-CB)
if DA<DB:
    print("point belongs to Centroid A")
else:
    print("point belongs to Centroid B")
```
